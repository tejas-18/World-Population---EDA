```python
asia_top10 = data[data["Continent"] == "Asia"][["Country/Territory", "2022 Population"]].sort_values(by="2022 Population",ascending=False, ignore_index=True)[:10]
```

```python
plt.figure(figsize=(12,4))
plt.title("Top 10 poulated country in Asia")
sns.barplot(data=asia_top10,x="Country/Territory",y="2022 Population",palette="Set1")
plt.show()
```

```python
for x in data["Continent"].unique():
    plt.figure(figsize=(12,4))
    temp = data[data["Continent"] == x][["Country/Territory", "2022 Population"]].sort_values(by="2022 Population",ascending=False, ignore_index=True)[:10]
    plt.title(f"Top 10 poulated country in {x}")
    sns.barplot(data=temp,x ="Country/Territory",y ="2022 Population",palette="Set1")
    plt.show()
```

```python
sns.countplot(x="Continent", data = data.sort_values(by="Continent",ascending=False))
```

```python
diff_70_22 = data.groupby("Continent")[["1970 Population","2022 Population"]].sum().sort_values(by="Continent",ascending=True).reset_index()
```

```python
diff_70_22
```

```python
plt.figure()
plt.plot(diff_70_22["Continent"],diff_70_22["1970 Population"],label="1970",color="red",marker="o")
plt.plot(diff_70_22["Continent"],diff_70_22["2022 Population"],label="2022",color="blue",marker="o")
plt.xlabel("Continent")
plt.ylabel("population")
plt.grid()
plt.legend()
plt.show()
```

```python
continent_data.sort_values(by = '2022 Population',inplace = True)
plt.figure(figsize=(12,9))
population_features = ['2022 Population','2020 Population','2015 Population','2010 Population','2000 Population','1990 Population']
population_features.reverse()
plt.figure(figsize=(15,6))
for feature in population_features:
    plt.plot(continent_data[feature],label = feature)
plt.legend()
plt.show()
```

```python
sorted_data = data.sort_values(by = "World Population Percentage",ascending=False)
```

```python
years = ["1970","1980","1990","2000","2010","2015","2020","2022"]
china_pop = [sorted_data[0:1]["1970 Population"],sorted_data[0:1]["1980 Population"],sorted_data[0:1]["1990 Population"],sorted_data[0:1]["2000 Population"],sorted_data[0:1]["2010 Population"],sorted_data[0:1]["2015 Population"],sorted_data[0:1]["2020 Population"],sorted_data[0:1]["2022 Population"]]
plt.step(years,china_pop)
plt.title("China Population Graph")
plt.show()
```
